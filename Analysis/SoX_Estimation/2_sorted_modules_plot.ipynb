```python
import pandas as pd
import os

path = './2st/'
df = pd.DataFrame()
for f in os.listdir(path):
    module = pd.read_excel(path+f, index_col = 0)
    df = pd.concat([df,module])
```

```python
df.to_csv("./2st_total_experiment_modules.csv")
```

```python
import matplotlib.pylab as plt
import numpy as np

# module별 개별 cell 전압, cell 최대/최소/평균/Gap 전압

yticks = [2.8, 3.0, 3.2, 3.4, 3.6, 3.8, 4.0, 4.2]

plt.rcParams['figure.figsize'] = [16, 10]
plt.rc('font', size=20) 

cmap = plt.get_cmap('jet_r')

cell_voltages = [ f'Cell_Voltage{i}' for i in range(1,13)]


for i in range(1,df.Module.max()+1):
    temp = df.loc[df['Module']==i]
    temp = temp[cell_voltages]
    
    mean = temp.mean(axis=1)
    maxx = temp.max(axis=1)
    minn = temp.min(axis=1)
    
    legend = f"Module {i} "
    print(f"Module {i}")
    #################################################################
    for c in range(len(temp.columns)):
        plt.subplot(2,2,1)

        plt.plot(temp.index[:-1], temp.iloc[:-1,c], linewidth=0.8, label = temp.columns[c], color = cmap(float(c)/12))

        plt.ylim([2.8,4.2])
        plt.yticks(yticks)
        plt.ylabel("Voltage(V)")
        
        print(c, temp.iloc[-2,c])
    
    plt.xticks(rotation=45)
    plt.legend(fontsize=10)
    ##################################################################
    t1 = pd.to_datetime('2022-04-06  19:47:42')
    t2 = pd.to_datetime('2022-04-06  20:45:00')
    
    for c in range(len(temp.columns)):
        plt.subplot(2,2,2)

        plt.plot(temp[t1:t2].index, temp.loc[t1:t2][temp.columns[c]], linewidth=0.8, label = temp.columns[c], color = cmap(float(c)/12))

        plt.ylim([2.8,3.4])
        plt.yticks([2.8,3.0,3.2])
        plt.ylabel("Voltage(V)")
    
    plt.xticks(rotation=45)
    plt.legend(fontsize=10)
    
    #################################################################
    plt.subplot(2, 2, 3) 
    
    plt.plot(temp.index, maxx, color = 'red', label = legend + "max")
    plt.plot(temp.index, mean, color = 'blue', label = legend + "mean")
    plt.plot(temp.index, minn, color = 'orange', label = legend + "min")
    
    plt.ylabel("Voltage(V)")
    plt.ylim([2.8, 4.1])
    
    plt.xticks(rotation=45)
    plt.yticks(yticks)
    
    plt.legend()
    
    ################################################################
    plt.subplot(2, 2, 4)
    
    plt.plot(temp.index, maxx-minn, color = 'red', linewidth=0.5, label = legend + "gap")
    
    plt.ylabel("Voltage(V)")
    plt.ylim([0,0.15])
    
    plt.xticks(rotation=45)
    plt.yticks([0, 0.03, 0.06, 0.09, 0.12, 0.15])
    
    plt.legend()
    
    #################################################################
    plt.tight_layout()
    plt.suptitle(legend + 'Voltages', fontsize=16)

    plt.savefig(f"./2st_fig/voltage_gap/{legend} Voltages.jpg", dpi=256)
    plt.show()
```

```python
import matplotlib.pylab as plt

# module별 cell 최대/최소/Gap 온도(Gap은 거의 없음)

plt.rcParams['figure.figsize'] = [16, 10]
plt.rc('font', size=20) 

for i in range(1,df.Module.max()+1):
    temp = df.loc[df['Module']==i]
    temp = temp[["MIN_CELL_TEMP","MAX_CELL_TEMP"]]
    
    maxx = temp.max(axis=1)
    minn = temp.min(axis=1)
    
    legend = f"Module {i} "
    ########################################################################################
    plt.subplot(2, 1, 1) 
    
    plt.plot(temp.index, maxx, color = 'red', label = legend + "max")
    plt.plot(temp.index, minn, color = 'orange', label = legend + "min")
    
    plt.ylabel("Temperature(℃)")
    plt.yticks([i for i in range(15,46,5)])
    plt.ylim([15,45])
    plt.legend()
    
    ########################################################################################
    plt.subplot(2, 1, 2)
    plt.plot(temp.index, maxx-minn, color = 'red', linewidth=0.5, label = legend + "gap")
    
    plt.ylabel("Temperature(℃)")
    plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.ylim([0,1])
    plt.legend()
    ########################################################################################
    plt.suptitle(legend + 'Temperatures', fontsize=32)
    
    plt.savefig(f"./2st_fig/temperature_gap/{legend} Temperatures.jpg", dpi=256)
    plt.show()
```
