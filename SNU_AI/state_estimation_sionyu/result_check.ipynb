```python
import torch
import os
import yaml

PATH = '/data/ess/output/State_Estimation/VITdV_SOC'
file_list = os.listdir(PATH)
file_list.sort()
# print(file_list[:10])

last = 50
minimum = 100
minimum_file = None
minimum_file_list = []
thres = 0.1

for i, file in enumerate(file_list[1:]):

    yaml_file = os.path.join(PATH, file, 'config.yaml')
    if os.path.exists(yaml_file):
        with open(yaml_file) as f:
            config = yaml.load(f, Loader=yaml.FullLoader)

            # get log
            if config['loss_fn'] == 'L1Loss' : 
                train_loss_log = torch.load( os.path.join(PATH, file, 'train_loss_log.pt') )
                val_loss_log   = torch.load( os.path.join(PATH, file, 'val_loss_log.pt') )
                train_min = min(train_loss_log[-last:])
                val_min = min(val_loss_log[-last:])
                
                if train_min <= thres and val_min <= thres :
                    minimum_file_list.append(file)
                    print(file, f'train_min = {train_min:.2f}', f'val_min = {val_min:.2f}')

                if val_min <= minimum :
                    minimum = val_min
                    minimum_file = os.path.join(PATH, file)

```

## Delete & Rename files

```python
import torch
import os
import yaml

PATH = '/data/ess/output/State_Estimation/VITdV_SOC'
file_list = os.listdir(PATH)
file_list.sort()
# print(file_list[:10])

last = 50
minimum = 100
minimum_file = None
minimum_file_list = []
thres = 0.1

for i, file in enumerate(file_list):

    yaml_file = os.path.join(PATH, file, 'config.yaml')
    if os.path.exists(yaml_file):
        with open(yaml_file) as f:
            config = yaml.load(f, Loader=yaml.FullLoader)

            # get log
            if config['loss_fn'] == 'L1Loss' : 
                train_loss_log = torch.load( os.path.join(PATH, file, 'train_loss_log.pt') )
                val_loss_log   = torch.load( os.path.join(PATH, file, 'val_loss_log.pt') )
                train_min = min(train_loss_log[-last:])
                val_min = min(val_loss_log[-last:])
                
                src = os.path.join(PATH, file)
                tar = os.path.join(PATH,f'train{train_min:.2f}_val{val_min:.2f}')
                if train_min <= thres and val_min <= thres :
                    minimum_file_list.append(file)
                    print(file, f'train_min = {train_min:.2f}', f'val_min = {val_min:.2f}')
                    shutil.move(src, tar) #rename
                    # os.rename(src, tar)
                else:
                    shutil.rmtree(src) # delete


                if val_min <= minimum :
                    minimum = val_min
                    minimum_file = os.path.join(PATH, file)

```
